### Gradient Descent Algorithm Implementation 

```python
import math
```

```python
def gradient_descent(x,y):
    m_curr = b_curr = 0
    iteration = 10000
    n = len(x)
    learning_rate = 0.0002
    prev_cost = float('inf') #initializing cost value higher than 1 to ensure intial if conditioin works
    
    for i in range(iteration):
        #y_predicted = mx+b
        #from linear equation y = mx+b 
        y_predicted = m_curr*x+b_curr

        #cost_function determines how well are we doing 
        #  at each step we should be reducing the cost
        #The most nearly the cost value difference with previous to zero the more accurate it is
        #If we find the optimim anwer is when the cost value donot change and remain same in each iteration
        cost = (1/n)*sum([val**2 for val in (y-y_predicted)])
        if math.isclose(prev_cost,cost,abs_tol=1e-5):
            print(f"conversion reched at iteration {i+1}")
            break
        
        md = -(2/n)*sum(x*(y-y_predicted)) #partial derivation of cost function(Mean Square Error) withrespect to m(coefficent)
        bd = -(2/n)*sum(y-y_predicted) #partial derivation of cost function(Mean Square Error) withrespect to b(intercept)

        m_curr = m_curr - learning_rate * md #gives the current point values after decrementing by learning rate and multiplying by slope 
        b_curr = b_curr - learning_rate * bd #gives the current point in the equation
        prev_cost = cost

        print(" m {}  b:{} Cost(C): {} iteration:{}".format(m_curr,b_curr,cost,i))

gradient_descent(x,y)
        
```
