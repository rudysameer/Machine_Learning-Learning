```python
#plotting the datapoints on a scatterplot chart to see wheather linear regression model can be applied
plt.scatter(car_df.Mileage,car_df['Sell Price($)'],label='cars',color='green',marker='*')
plt.title('Scatter Plot for The Car')
plt.xlabel('Mileage in km')
plt.ylabel('prices')
plt.legend()
plt.show()
#seeing the plot we can conclude we can do Linear regression as it is closely being in linear form
```

### Building a Model to To Predict the Following quesions 
- Since there is different car models we neeed to make dummies variables for them to uniquely identify them
- 

```python
#Using pd.get_dummies for creating dummies
dummies = pd.get_dummies(car_df['Car Model'])
#dummies we got in boolean(True/False) changing it to the integer
dummies = dummies.astype(int)
dummies
```

```python
#droppping one of the dummy variable to avoid variable colinarity which can cause Dummy variable Problem.
dummies = dummies.drop(['Mercedez Benz C class'],axis = 1)
dummies
```

```python
#merging with the car_df
car_df = pd.concat([car_df,dummies],axis=1) 
car_df
```

```python
#we donot need name of the car sa we have already made dummy variable to determine them
#dropping Car Model

#car_df = car_df.drop(['Car Model'],axis=1)
car_df
```

```python
car_df #car_model dropped succesfully but the columns are duplicated may be i ran twice the same program so dropping them
car_df = car_df.loc[:,~car_df.columns.duplicated()]
car_df
```

### Modelling Linear Regression
- dependent variable is Sell Price($)
- independent Variable is Mileage,Age,audi,BMW


```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(car_df[['Mileage','Age(yrs)','Audi A5','BMW X5']],car_df['Sell Price($)'])
```

### Predicting the Question
- Predict price of a mercedez benz that is 4 yr old with mileage 45000- 
Predict price of a BMW X5 that is 7 yr old with mileage 86000-
- 
Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())

```python
#Predict price of a mercedez benz that is 4 yr old with mileage 45000
model.predict([[45000,4,0,0]]) #answer is close
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86000
model.predict([[86000,7,0,1]])
```

**model accuracy**

```python
#- Tell me the score (accuracy) of your model. (Hint: use LinearRegression().score())
model.score(car_df[['Mileage','Age(yrs)','Audi A5','BMW X5']],car_df['Sell Price($)']) #its 94 % accurate based on the actual data from the datapoints
```

## Using One Hot Encoding

```python
OheCar = pd.read_csv(r'csvs/carprices.csv')
OheCar
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
#initializing the class object
ohe = OneHotEncoder()
#fitting and transforming the category column('Car Model')
encoded_features = ohe.fit_transform(OheCar[['Car Model']]).toarray()
encoded_features #gives dummy variables
```

```python
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out(['Car Model']))
encoded_df
```

```python
#replacing the column names 
# Replace all column names
encoded_df.columns = ['Audi A5','BMW X5','Mercedez']
encoded_df
```

```python
final_df_ohe = pd.concat([OheCar,encoded_df],axis=1).drop(['Mercedez','Car Model'],axis=1)
```

```python
final_df_ohe
```
