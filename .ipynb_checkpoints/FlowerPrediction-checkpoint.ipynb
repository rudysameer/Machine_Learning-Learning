### STEP 1: Converting data into dataframe for smooth operation

```python
import pandas as pd
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
iris_df.head()
```

```python
iris_df['target'] = iris.target
```

```python
iris_df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target']
```

```python
iris_df[100:125]
```

```python
#renaming columns
```

### Step2 Training and testing data Split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width',]],iris_df['target'],test_size=0.2)
```

```python
#splitted sucessfully
print(f"Training data: {x_train.shape} , Training Target Data {y_train.shape}")
print(f"Testing data: {x_test.shape} , Testing Target Data {y_test.shape}")
```

### Creating Model
- plotting the data points
- Building model using Logistic Regression
- 

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(iris_df, hue="target", palette="Set1")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

```python
#showing names insted of number
iris_df['species'] = iris_df['target'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
sns.pairplot(iris_df, hue="species", palette="Set1")
plt.suptitle("Pairplot of Iris Dataset", y=1.02)
plt.show()
```

### Training Model using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
iris_df[50:90]
```

```python
prediction_df = iris_df[['sepal_length','sepal_width','petal_length','petal_width']]
prediction_df
```

```python
answer_predicted = model.predict(prediction_df)
```

```python
answer_predicted[50:145]
```

```python
model.predict(x_test)
```

### Creating Confusion Matix to predict overall Performance of the model

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(iris_df.target,answer_predicted)
```

```python
cm
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(cm,annot=True)

plt.xlabel("Predicted")
plt.ylabel("Actual value")
print("0='setosa', 1='versicolor, 2='Virginica' ")
plt.show()
```
