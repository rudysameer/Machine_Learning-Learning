```python
import pandas as pd
```

```python
df = pd.read_csv(r'csvs/homeprices.csv')
```

```python
df
```

- The new thing here is **Town** .Now the predication prices have to be different based on different areas
- There are two different types of categories **Nominal Variables** and **Ordinal Variables**
- **Nominal Variables** doot have order or number relationship beetween eachother ie gender,color,places etc categories 
- **Ordinal Variables** have some sort of numerical ordering or hieraary beetween each other. i.e Education level, (high,medium,low),(satsfied,nuetral,dissatisfied). We can Order it numerically nad use simple encoding replacing high=3,medium=2,low=1
- In **Nominal categories** we Cannot Use simple Encoding. We Use Concept of **Dummy Variables**

```python
print(df.town.dtypes)
```

```python
#creating dummy varibles based on the town name
dummies = pd.get_dummies(df.town)
dummies
```

```python
print(dummies.dtypes)
#since the dtype of the town is bool so it is giving true/false.Convert it to int
```

```python
#converting bool into integer
dummies = dummies.astype(int)
dummies
```

```python
#merging dummy varibles with dataframe
merged = pd.concat([df,dummies],axis='columns')
merged
```

- Now we have to **drop town**. Since we have dummy varibles to define them.
- We have to drop one of the dummy varibles to avoid **Dummy Variable Trap**.
- When each variable can we derived from rest of the variable it is called **multi-co-linear**. Which creates **dummy variable trap** 

```python
#dropping town and west windsor
final = merged.drop(['town','west windsor'],axis='columns')
final
```

- We would be using LinearRegression Model which requires **independent variable** and **dependent variable**
- model.fit(independent_vaiable,dependent_variable)
- **independent_variable** are area,monroe,robinsville
- **dependent_variable** price

```python
independent_variable = final.drop('price',axis='columns')
independent_variable
```

```python
dependent_variable = final.price
dependent_variable
```

### Modelling LinearRegression for Prediction


```python
from sklearn import linear_model
```

```python
model = linear_model.LinearRegression()
model.fit(independent_variable,dependent_variable)
```

### Now we can predict the prices according to the places
- Predict price for 2800sqft in robinsvilla
- predict price for 3400sqft in west windsor

```python
#predicting forfor 2800sqft in robinsvilla
#model.predict([[area,monroe,robinsvilla]])
model.predict([[2800,0,1]])
```

```python
#predictiong for 3400sqft in west windsor
model.predict([[3400,0,0]])
```

```python
#if we want to know how accurate is our model the we use model.score(x,y)
#it will calculate predicted values for all of the rows in x, and compare with actual values which is y
model.score(independent_variable,dependent_variable)
#its 95.739% acurate
```

## We would do Same thing using Different Method called One Hot Encoding library

- In order to use **One Hot Encoder** we need to label encoding on town column

```python
#label Encoding on town column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder() #create label Encoder class object
```

```python
# making new df for predicting
dfle = df
dfle
```

```python
#use label Encoder to fit and transform
dfle.town = le.fit_transform(dfle.town) #it takes label column as input and returns the label array. Converts label into integer
dfle
```

 once label is encoded we need to make **x(independent variable) and y(dependent variable)** for **model.fit(x,y)**

```python
independent = dfle[['town','area']].values #gives 2d array instead of dataframe
independent
```

```python
dependent = dfle.price
dependent
```

#### We know we need to Create Dummy Variables.
#### For That we use onehotencoder libarry

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder() #making an OneHotEcoder class library
# Fit and transform the 'town' column
encoded_features = ohe.fit_transform(df[['town']]).toarray()
encoded_features# it made dummy variables with respect to town columns
```

```python
encoded_df = pd.DataFrame(encoded_features, columns=ohe.get_feature_names_out(['town']))
#This line creates a pandas DataFrame from the encoded features produced by the OneHotEncoder, with appropriate column names for each generated feature.
encoded_df
```

```python
new_df = pd.concat([encoded_df,dfle],axis='columns').drop('town',axis=1)
new_df
#it made dummy variables 
```

```python
#dropping one varables to avoid Dummy Variable trap
new_df = new_df.drop('town_2',axis=1)
new_df
```

### Creating Linear Regression Model


```python
ohmodel = linear_model.LinearRegression()
ohmodel.fit(new_df[['town_0','town_1','area']],new_df.price)
```

```python
#predictiong for 3400sqft in west windsor
ohmodel.predict([[0,0,3400]])
```
