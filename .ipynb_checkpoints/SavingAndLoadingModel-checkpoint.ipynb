```python
import pandas as pd
```

```python
area_df = pd.read_csv('AreaPricePrediction.csv')
```

```python
area_df
```

### Training a Model from DataSet 

```python
from sklearn import linear_model 
```

```python
lmodel = linear_model.LinearRegression()
lmodel.fit(area_df[['area']],area_df.Predicted_Prices)
```

```python
lmodel.predict([[6000]])
```

```python
print(lmodel.coef_)
lmodel.intercept_
```

### Saving a Trained Model to a File
- so we donot have to train using dataset again and again

```python
#using pickle from sklearn
import pickle
```

```python
with open('model_pickle','wb')as f: # 'model_pickle' is the file name where we want to save the file
    #'wb' as writing in binary
    pickle.dump(lmodel,f)#dumpig the lmodel we trained into the file 
    #it created 'model_pickle' in my working directory which have gibberish words as it is in binary 
```

### Importing the Same model
- We can Use it without training dataset

```python
with open('model_pickle','rb') as f: #'model_pickle' file name we want to import
    #'rb' as reding the file in binary as it is in binary
    import_model = pickle.load(f)
    
    
```

Now we Have a Model name 'import_model' loaded from a file into a memory. 'import_model' as object.
- Using it to generate answers wheather it is same as above

```python
print(import_model.predict([[6000]]))
print(lmodel.predict([[6000]]))
#gives the same answer
```

```python
#for coefficent and intercept 
print('using trained dataset model')
print(lmodel.coef_)
print(lmodel.intercept_)

print('using imported loaded model')
print(import_model.coef_)
print(import_model.intercept_)
```

### Second Apporch: sklearn Joblib
- This model is bascically same as Pckle but it is more effecient when there is lots of numpy arrays in the data set as per professionals

```python
import joblib
```

```python
joblib.dump(lmodel,'Joblibmodel')#it made new file joblibmodel in the working directory
```

```python
jobModel = joblib.load('Joblibmodel')
```

```python
print('Using Joblig')
print(jobModel.predict([[6000]]))
print(jobModel.coef_)
print(jobModel.intercept_)
```
